Tiến hành thu thập dữ liệu từ trang Wikipedia về các quốc gia ASEAN theo GDP. Kiểm tra số lượng bảng, tiến hành cạo dữ liệu hai bảng 4, bảng 5 và tiến hành lưu file.

```r
# Cài đặt thư viện nếu chưa có
  if (!require("rvest")) install.packages("rvest", dependencies = TRUE)

  # Load thư viện
  library(rvest)

  # URL trang Wikipedia
  url <- "https://en.wikipedia.org/wiki/List_of_ASEAN_countries_by_GDP"

  # Đọc HTML của trang
  page <- read_html(url)

  # Lấy tất cả các bảng trên trang
  tables <- page %>% html_elements("table.wikitable")

  # Kiểm tra số lượng bảng
  cat("📌 Tổng số bảng tìm thấy:", length(tables), "\n")

  # Nếu có ít nhất 5 bảng thì lấy bảng 4 và bảng 5
  if (length(tables) >= 5) {
    # Đọc bảng 4
    table_4 <- tables[[4]] %>% html_table(fill = TRUE) %>% as.data.frame()
    cat("\n📌 Bảng 4 - Dữ liệu đầy đủ:\n")
    print(table_4)

    # Đọc bảng 5
    table_5 <- tables[[5]] %>% html_table(fill = TRUE) %>% as.data.frame()
    cat("\n📌 Bảng 5 - Dữ liệu đầy đủ:\n")
    print(table_5)

    # Lưu dữ liệu vào file CSV
    write.csv(table_4, "Bang_4.csv", row.names = FALSE)
    write.csv(table_5, "Bang_5.csv", row.names = FALSE)

    cat("\n✅ Dữ liệu đã được lưu vào 'Bang_4.csv' và 'Bang_5.csv'\n")
  } else {
    cat("❌ Không tìm thấy đủ 5 bảng trên trang!\n")
  }
```

```r
  # Load thư viện
  library(tidyverse)

  library(tidyverse)

  # Đọc dữ liệu từ các file CSV
  df_2020_2021 <- read_csv("gdp_asean_2020_2021_fixed.csv") %>%
    select(Quốc_gia, GDP_2020, GDP_2021)

  df_2022 <- read_csv("GDP_ASEAN_Bang4_Filtered.csv") %>%
    select(Quốc_gia = Country, GDP_2022 = GDP_Nominal_per_Capita_USD)

  df_2023 <- read_csv("GDP_ASEAN_Bang5_Filtered.csv") %>%
    select(Quốc_gia = Country, GDP_2023 = GDP_Nominal_per_Capita_USD)

  # 🛠 Chuẩn hóa tên quốc gia để tránh trùng lặp
  df_2022$Quốc_gia <- recode(df_2022$Quốc_gia,
                             "Cambodia" = "Campuchia",
                             "Laos" = "Lào",
                             "Thailand" = "Thái Lan",
                             "Vietnam" = "Việt Nam")

  df_2023$Quốc_gia <- recode(df_2023$Quốc_gia,
                             "Cambodia" = "Campuchia",
                             "Laos" = "Lào",
                             "Thailand" = "Thái Lan",
                             "Vietnam" = "Việt Nam")

  # Gộp dữ liệu & loại bỏ ASEAN
  df_merged <- reduce(list(df_2020_2021, df_2022, df_2023), full_join, by = "Quốc_gia") %>%
    filter(Quốc_gia != "ASEAN")

  # 🛠 Xử lý giá trị NA (có thể thay = 0 hoặc lấy trung bình)
  df_merged <- df_merged %>%
    mutate(across(starts_with("GDP_"), ~ replace_na(., 0)))

  # 🖥 In dữ liệu hoàn chỉnh ra màn hình
  print(df_merged, n = Inf)

  # Lưu file CSV kết quả
  write_csv(df_merged, "GDP_ASEAN_2020_2023_Cleaned.csv")
  print("✅ File GDP_ASEAN_2020_2023_Cleaned.csv đã lưu thành công!")

```

*   Tính tỷ lệ tăng trưởng trung bình giữa hai năm.

Cập nhật GDP_2022 và GDP_2023:
  Nếu giá trị GDP của Đông Timor trong năm 2022 hoặc 2023 là 0, thì sẽ dự đoán giá trị này theo tỷ lệ tăng trưởng từ 2020 đến 2021.

Làm tròn kết quả: Làm tròn giá trị GDP của các năm để loại bỏ phần thập phân.




```r
# Tính tỷ lệ tăng trưởng trung bình của Đông Timor từ 2020 → 2021
  gdp_dt_2020 <- df_merged$GDP_2020[df_merged$Quốc_gia == "Đông Timor"]
  gdp_dt_2021 <- df_merged$GDP_2021[df_merged$Quốc_gia == "Đông Timor"]

  tang_truong_tb <- (gdp_dt_2021 - gdp_dt_2020) / gdp_dt_2020

  # Cập nhật GDP_2022 và GDP_2023 nếu có giá trị 0 bằng dự đoán theo tỷ lệ tăng trưởng
  df_merged <- df_merged %>%
    mutate(
      GDP_2022 = ifelse(Quốc_gia == "Đông Timor" & GDP_2022 == 0, gdp_dt_2021 * (1 + tang_truong_tb), GDP_2022),
      GDP_2023 = ifelse(Quốc_gia == "Đông Timor" & GDP_2023 == 0, GDP_2022 * (1 + tang_truong_tb), GDP_2023)
    ) %>%
    # Làm tròn số để loại bỏ dấu thập phân
    mutate(across(starts_with("GDP"), round, digits = 0))

  # In lại toàn bộ bảng dữ liệu sau khi cập nhật
  print(df_merged, n = Inf)

  # Lưu file CSV kết quả
  write_csv(df_merged, "GDP_ASEAN_2020_2023_Cleaned.csv")
  print("✅ File GDP_ASEAN_2020_2023_Cleaned.csv đã lưu thành công!")

```

```r
library(tidyverse)

  # Đọc dữ liệu từ các file CSV
  df_gdp <- read_csv("GDP_ASEAN_2020_2023_Cleaned.csv")
  df_dan_so <- read_csv("dan_so_2020_2023_fixed.csv") %>% rename(Quốc_gia = Country)
  df_that_nghiep <- read_csv("ty_le_that_nghiep_cleaned.csv") %>% rename(Quốc_gia = `Quốc gia`)

  # Gộp dữ liệu theo "Quốc_gia"
  df_merged <- df_dan_so %>%
    full_join(df_gdp, by = "Quốc_gia") %>%
    full_join(df_that_nghiep, by = "Quốc_gia")

  # Loại bỏ dòng trùng lặp hoặc có toàn bộ giá trị NA
  df_merged <- df_merged %>% distinct(Quốc_gia, .keep_all = TRUE) %>% drop_na(Quốc_gia)

  # Xóa các cột có toàn bộ giá trị NA
  df_merged <- df_merged %>% select(where(~ !all(is.na(.))))

  # Sắp xếp lại thứ tự cột: Dân số → GDP → Tỷ lệ thất nghiệp
  df_merged <- df_merged %>%
    select(Quốc_gia, starts_with("Dan_so"), starts_with("GDP"), `2020`, `2021`, `2022`, `2023`)

  # Xuất file CSV kết quả
  write_csv(df_merged, "ASEAN_2020_2023_Final.csv")

  # In kết quả kiểm tra
  print(df_merged, n = Inf)
  print("✅ File ASEAN_2020_2023_Final.csv đã lưu thành công!")
```

```r
library(DT)
# Hàm tính tốc độ tăng trưởng cho một cột
calculate_growth <- function(df, prefix, years) {
  for (i in 2:length(years)) {
    year_prev <- years[i - 1]
    year_curr <- years[i]

    # Tạo tên cột cho năm trước và năm hiện tại
    col_prev <- paste0(prefix, year_prev)
    col_curr <- paste0(prefix, year_curr)

    # Kiểm tra xem các cột có tồn tại không
    if (!col_prev %in% colnames(df) || !col_curr %in% colnames(df)) {
      stop(paste("Cột không tồn tại:", col_prev, col_curr))
    }

    # Tạo cột tốc độ tăng trưởng
    growth_col <- paste0("Tang_truong_", prefix, year_curr)

    # Tính tốc độ tăng trưởng và thêm vào dataframe
    df[[growth_col]] <- (df[[col_curr]] - df[[col_prev]]) / df[[col_prev]] * 100
  }
  return(df)
}

# Đọc dữ liệu từ tệp CSV
file_path <- "ASEAN_2020_2023_Final_Cleaned.csv"
df <- read.csv(file_path)

# Danh sách các năm
years <- c(2020, 2021, 2022, 2023)

# Kiểm tra xem các cột có tồn tại không
if (!all(c("Dan_so_2020", "Dan_so_2021", "Dan_so_2022", "Dan_so_2023",
           "GDP_2020", "GDP_2021", "GDP_2022", "GDP_2023",
           "X2020", "X2021", "X2022", "X2023") %in% colnames(df))) {
  stop("Một trong các cột không tồn tại trong dataframe.")
}

# Tính tốc độ tăng trưởng cho dân số, GDP và tỷ lệ thất nghiệp
df <- calculate_growth(df, "Dan_so_", years)      # Dân số
df <- calculate_growth(df, "GDP_", years)          # GDP
df <- calculate_growth(df, "X", years)             # Thất nghiệp (dữ liệu là cột X2020, X2021, v.v.)

# Kiểm tra các cột tăng trưởng đã được tạo chưa
print(colnames(df))

# Hiển thị toàn bộ dữ liệu cho tất cả các quốc gia và các chỉ số tăng trưởng
# In toàn bộ bảng dữ liệu
print(df)

# Nếu bạn muốn lưu dữ liệu vào tệp CSV sau khi tính toán tốc độ tăng trưởng
output_file_path <- "ASEAN_tang_truong_2020_2023.csv"
write.csv(df, output_file_path, row.names = FALSE)

# Kiểm tra xem tệp đã được lưu thành công chưa
cat("Dữ liệu đã được lưu vào tệp:", output_file_path)
```

##Kiểm định Shapiro-Wilk, Biểu đồ mật độ (Density Plot)



**1. Kiểm định Shapiro-Wilk**


*   Mục đích: Kiểm tra xem một tập dữ liệu có tuân theo phân phối chuẩn hay không.
*   So sánh phân phối thực tế của dữ liệu với phân phối chuẩn dự kiến.

*   Giá trị p:
   *   p>0.05: Dữ liệu có thể coi là tuân theo phân phối chuẩn.
   *   p≤0.05: Dữ liệu không tuân theo phân phối chuẩn.










**2. Biểu đồ mật độ (Density Plot)**

*   Mục đích: Hiển thị phân bố của dữ liệu dưới dạng hàm mật độ xác suất (PDF).

*   Đặc điểm:
   *   Mịn hơn so với histogram.

   *   Giúp quan sát xu hướng phân bố, đối xứng và đa đỉnh.

*   Ứng dụng: Kiểm tra trực quan tính chuẩn của dữ liệu trước khi thực hiện kiểm định thống kê.





```r
  library(tidyverse)

  # Đọc dữ liệu
  df <- read_csv("ASEAN_2020_2023_Final_Cleaned.csv")

  # Kiểm tra tên cột thực sự
  print(colnames(df))

  # Thực hiện kiểm định Shapiro-Wilk cho từng cột số liệu
  shapiro_results <- df %>%
    select(-`Quốc_gia`) %>%  # Bỏ cột tên quốc gia
    pivot_longer(everything(), names_to = "Variable", values_to = "Value") %>%
    group_by(Variable) %>%
    summarise(P_value = shapiro.test(Value)$p.value)

  # In kết quả
  print(shapiro_results)
```

```r
library(tidyverse)
shapiro_results <- df %>%
  select(-Quốc_gia) %>%  # Bỏ cột tên quốc gia
  pivot_longer(everything(), names_to = "Variable", values_to = "Value") %>%
  group_by(Variable) %>%
  summarise(
    Sample_Size = sum(!is.na(Value)),  # Kiểm tra số lượng quan sát
    P_value = ifelse(Sample_Size > 3 & Sample_Size <= 5000,
                     shapiro.test(na.omit(Value))$p.value,
                     NA)  # Trả về NA nếu mẫu không hợp lệ
  )

# In kết quả
print(shapiro_results)
```

1. Trung bình (Mean)

Định nghĩa:
+ Trung bình (Mean) là giá trị đại diện cho một tập hợp số liệu, được tính bằng cách cộng tất cả các giá trị lại rồi chia cho số lượng phần tử.

Ưu điểm:
+ Dễ tính toán và sử dụng.
+ Phù hợp khi dữ liệu không có giá trị ngoại lai.

Nhược điểm:
+ Dễ bị ảnh hưởng bởi các giá trị ngoại lai (outliers). Ví dụ, nếu một quốc gia có GDP cực cao, trung bình sẽ bị đẩy lên cao hơn thực tế.


2. Trung vị (Median)

Định nghĩa:

+ Trung vị (Median) là giá trị nằm ở vị trí chính giữa khi sắp xếp dữ liệu theo thứ tự tăng dần hoặc giảm dần.

Cách tính:

+ Nếu số lượng phần tử lẻ: Trung vị là giá trị ở giữa.
+ Nếu số lượng phần tử chẵn: Trung vị là trung bình của hai giá trị ở giữa.

Ưu điểm:
+ Không bị ảnh hưởng bởi giá trị ngoại lai.
+ Phù hợp khi dữ liệu không đối xứng.

Nhược điểm:
+ Không phản ánh chính xác giá trị tổng thể như trung bình.

## BIỂU ĐỒ HEATMAP(NHUNG)

Cài đặt và tải các thư viện cần thiết để phân tích và trực quan hóa dữ liệu, đặc biệt là vẽ biểu đồ tương quan.

*   corrplot: Hỗ trợ vẽ biểu đồ ma trận tương quan.
*   reshape2: Hỗ trợ chuyển đổi dữ liệu thành định dạng phù hợp cho phân tích.



```r
# Cài đặt và tải các thư viện cần thiết
  install.packages("corrplot", dependencies = TRUE)  # Cài đặt corrplot nếu chưa có
  install.packages("reshape2", dependencies = TRUE)  # Cài đặt reshape2 nếu chưa có

  library(tidyverse)
  library(reshape2)
  library(corrplot)
```

Phân tích mối tương quan giữa các biến số (GDP, dân số, tỷ lệ thất nghiệp) của ASEAN từ 2020-2023 bằng cách tính ma trận tương quan và trực quan hóa bằng biểu đồ heatmap.

*   Chuyển đổi dữ liệu sang dạng số (trừ cột "Quốc_gia").
*   Tính toán ma trận tương quan giữa các biến số.

*   Vẽ heatmap để hiển thị mức độ tương quan giữa các biến.

Biểu đồ heatmap giúp quan sát trực quan các mối quan hệ mạnh/yếu giữa GDP, dân số, và tỷ lệ thất nghiệp.




```r
# Đọc dữ liệu
  df <- read_csv("ASEAN_2020_2023_Final_Cleaned.csv")

  # Chuyển đổi dữ liệu thành dạng số nếu chưa đúng
  df_numeric <- df %>%
    select(-Quốc_gia) %>%  # Loại bỏ cột quốc gia (nếu có)
    mutate(across(everything(), as.numeric))

  # Tính ma trận tương quan
  cor_matrix <- cor(df_numeric, use = "complete.obs")

  # Vẽ heatmap (bản đồ nhiệt)
  corrplot(cor_matrix, method = "color", type = "upper", tl.col = "black", tl.srt = 45)
```

Nhận xét về biểu đồ tương quan

## Kiểm định Kolmogorov-Smirnov (K-S Test)
# Các tài liệu tham khảo về kiểm định
(https://www.itl.nist.gov/div898/handbook/eda/section3/eda35g.htm)
(https://dataaspirant.com/kolmogorov-smirnov-test/)
# Giới thiệu
- Kiểm định Kolmogorov-Smirnov (K-S Test) là một kiểm định phi tham số để so sánh sự tương đồng giữa hai phân phối xác suất. Nó được sử dụng phổ biến để:
  + Kiểm tra xem một mẫu dữ liệu có tuân theo một phân phối lý thuyết nhất định hay không.
  + So sánh hai mẫu dữ liệu xem chúng có đến từ cùng một phân phối hay không.
- Ưu điểm:
  + Không yêu cầu giả định về dạng phân phối của dữ liệu.
  + Có thể áp dụng cho cả mẫu nhỏ và lớn.
- Nhược điểm:
  + Nhạy cảm với sự khác biệt ở trung tâm của phân phối hơn là ở hai đuôi.
  + Khi dữ liệu có nhiều điểm lặp lại, kiểm định có thể không chính xác.

# 2.2 Giả thuyết kiểm định
- H₀: Dữ liệu tuân theo phân phối chuẩn.
- H₁: Dữ liệu không tuân theo phân phối chuẩn

# Thu thập dữ liệu
# Kiểm định K-S cho từng biến
# Hiển thị kết quả kiểm định
# Vẽ biểu đồ Q-Q Plot để kiểm tra trực quan
# Nhận xét về phân phối dữ liệu
- Nếu giá trị p-value < 0.05, ta bác bỏ giả thuyết H₀, tức là dữ liệu không tuân theo phân phối chuẩn.
- Nếu giá trị p-value > 0.05, ta không bác bỏ H₀, có thể kết luận dữ liệu có phân phối chuẩn.
- Kết hợp với Q-Q Plot:
  + Nếu các điểm nằm gần đường chéo, dữ liệu có xu hướng tuân theo phân phối chuẩn.
  + Nếu có nhiều điểm lệch xa đường chéo, dữ liệu có thể có phân phối lệch hoặc chứa outliers.

# Kết luận kiểm định
- Kết quả kiểm định K-S cho thấy biến Dân số, GDP và Tỷ lệ thất nghiệp có p-value < 0.05, nghĩa là các biến này không tuân theo phân phối chuẩn.
- Vì vậy, khi phân tích dữ liệu, ta nên sử dụng các phương pháp kiểm định phi tham số.

## BIỂU ĐỒ BOXPLOT(TRANG)

- Biểu đồ Boxplot này thể hiện sự phân bố GDP của các quốc gia ASEAN trong giai đoạn 2020 - 2023, khi các quốc gia được phân loại theo quy mô dân số (nhỏ, trung bình, lớn) dựa trên khoảng phân vị
- Để chia dữ liệu thành ba nhóm, ta sử dụng phân vị (quantiles) như sau:

 + Phân vị thứ nhất (Q1 - 33.3%): Điểm cắt giữa nhóm dân số nhỏ và trung bình.

 + Phân vị thứ hai (Q2 - 66.6%): Điểm cắt giữa nhóm dân số trung bình và dân số lớn
- Ý nghĩa của việc chia nhóm theo khoảng phân vị dân số: Thay vì xem xét từng quốc gia riêng lẻ, các nước ASEAN được chia thành ba nhóm dân số dựa trên các phân vị thống kê:

 + Nhóm dân số nhỏ (NHO - Màu đỏ): Bao gồm các quốc gia có dân số thuộc phân vị thấp nhất.

 + Nhóm dân số trung bình (TRUNG BÌNH - Màu xanh lá): Bao gồm các quốc gia có dân số nằm trong phân vị giữa.

 + Nhóm dân số lớn (LON - Màu xanh dương): Bao gồm các quốc gia có dân số thuộc phân vị cao nhất.

Mục đích của cách phân chia này là để xem mối quan hệ giữa GDP và quy mô dân số có tuân theo một xu hướng nhất quán hay không.

```r
# Cài đặt và nạp thư viện nếu chưa có
install.packages("tidyverse", repos = "http://cran.rstudio.com/")
library(tidyverse)
library(ggplot2)

# Đọc dữ liệu từ file CSV đã tải lên Colab
df_gdp <- read_csv("GDP_ASEAN_2020_2023_Cleaned.csv")
df_danso <- read_csv("dan_so_2020_2023_fixed.csv")

# Chuyển dữ liệu từ dạng wide -> long
df_gdp_long <- df_gdp %>%
  pivot_longer(cols = starts_with("GDP"), names_to = "Nam", values_to = "GDP") %>%
  rename(Quoc_gia = `Quốc_gia`)

df_danso_long <- df_danso %>%
  pivot_longer(cols = starts_with("Dan_so"), names_to = "Nam", values_to = "Dan_so") %>%
  rename(Quoc_gia = Country)

# Chuẩn hóa cột năm để khớp giữa hai dataset
df_gdp_long <- df_gdp_long %>% mutate(Nam = gsub("GDP_", "", Nam))
df_danso_long <- df_danso_long %>% mutate(Nam = gsub("Dan_so_", "", Nam))

# Gộp hai bộ dữ liệu theo quốc gia và năm
df_merged <- inner_join(df_gdp_long, df_danso_long, by = c("Quoc_gia", "Nam"))

# Kiểm tra dữ liệu đầu vào
head(df_merged)

# Xác định các khoảng phân vị (loại bỏ giá trị trùng)
breaks <- unique(quantile(df_merged$Dan_so, probs = c(0, 1/3, 2/3, 1), na.rm = TRUE))

# Kiểm tra xem có đúng 3 khoảng không
if (length(breaks) == 4) {
  df_merged <- df_merged %>%
    mutate(Nhom_dan_so = cut(Dan_so,
                             breaks = breaks,
                             labels = c("NHO", "TRUNG BINH", "LON"),
                             include.lowest = TRUE))
  print("Phân nhóm thành công!")
} else {
  print("Lỗi: Không thể tạo đúng 3 nhóm dân số. Hãy kiểm tra dữ liệu!")
}

# Kiểm tra kết quả
table(df_merged$Nhom_dan_so)

# Vẽ biểu đồ Boxplot phân bố GDP theo nhóm dân số
ggplot(df_merged, aes(x = Nhom_dan_so, y = GDP, fill = Nhom_dan_so)) +
  geom_boxplot() +
  theme_minimal() +
  labs(title = "PHAN BO GDP THEO NHOM DAN SO ASEAN (2020 - 2023)",
       x = "NHOM DAN SO", y = "GDP (TY USD)") +
  theme(legend.position = "none")
```

Nhận xét về biểu đồ Boxplot phân bố GDP theo nhóm dân số (2020 - 2023)

- Nhóm dân số nhỏ (NHO - Màu đỏ)
 + GDP có sự phân tán lớn nhất, thể hiện qua hộp (IQR) có độ dài rộng .

 + GDP trung vị cao hơn so với hai nhóm còn lại, điều này có thể do một số quốc gia có dân số nhỏ nhưng GDP lớn (ví dụ: Singapore, Brunei).

 + Có nhiều giá trị ngoại lai (outliers), có thể là những quốc gia có GDP cực cao so với các nước còn lại trong nhóm này.

- Nhóm dân số trung bình (TRUNG BÌNH - Màu xanh lá)
 + GDP có phạm vi thấp hơn đáng kể so với nhóm dân số nhỏ.

 + Trung vị GDP thấp, cho thấy hầu hết các nước trong nhóm này có mức GDP không cao.

 + IQR nhỏ, tức là GDP của các quốc gia trong nhóm này khá đồng đều.

- Nhóm dân số lớn (LON - Màu xanh dương)
 + GDP tập trung trong khoảng rất hẹp, với whiskers ngắn.

 + Trung vị GDP cũng thấp, cho thấy các quốc gia có dân số lớn chưa chắc có GDP cao.

 + Một số outliers có thể là các quốc gia có dân số lớn nhưng GDP vượt trội (ví dụ: Indonesia).

- Kết luận
 + GDP không tỉ lệ thuận với dân số : Nhóm dân số nhỏ có GDP biến động lớn và cao hơn nhóm dân số lớn.
 + Nhóm dân số trung bình có GDP đồng đều hơn, nhưng trung bình thấp hơn nhóm nhỏ.
 + Nhóm dân số lớn có GDP khá thấp và ít biến động, có thể do nhiều quốc gia có dân số lớn nhưng GDP chưa cao.

# Hệ số tương quan
- Hệ số tương quan đo mức độ và chiều hướng liên hệ giữa hai biến:

  - Dao động từ -1 đến 1.

     + Gần 1 → tương quan dương mạnh.
     + Gần -1 → tương quan âm mạnh.
     + Gần 0 → gần như không có mối tương quan.

- Do dữ liệu không tuân theo phân phối chuẩn (đã được kiểm định trước đó), nhóm sử dụng hệ số tương quan Spearman để đánh giá mức độ và chiều hướng tương quan giữa các biến.

- GDP & Thất nghiệp:	+0.73 ->	Tương quan dương mạnh → GDP cao thường đi kèm với tỷ lệ thất nghiệp cao hơn trong giai đoạn này (có thể do ảnh hưởng COVID).

# Độ đồng biến (Covariance)
- Độ đồng biến thể hiện xu hướng cùng thay đổi giữa hai biến:
   + Nếu cov > 0: hai biến cùng chiều (biến này tăng → biến kia tăng).
   + Nếu cov < 0: hai biến ngược chiều.

- Khác với hệ số tương quan, độ đồng biến có đơn vị và phụ thuộc vào quy mô dữ liệu.


-Nhận xét kỳ vọng:
  + Dân số – GDP: Đồng biến dương nhẹ, GDP tăng khi dân số tăng (nhưng không ổn định).

  + Dân số – Thất nghiệp: Đồng biến âm nhẹ, cho thấy dân số không ảnh hưởng lớn đến thất nghiệp.

*   Màu xanh đậm: Tương quan âm mạnh (giá trị gần -1) → Khi một biến tăng, biến kia giảm.

*   Nếu GDP và dân số có tương quan dương, nghĩa là khi dân số tăng, kinh tế cũng phát triển.
*   Nếu dân số và tỷ lệ thất nghiệp có tương quan âm, có thể thấy khi dân số tăng, tỷ lệ thất nghiệp giảm do thị trường lao động mở rộng.

Tóm lại, heatmap giúp chúng ta có cái nhìn tổng quan về sự liên kết giữa các yếu tố kinh tế - xã hội, từ đó đưa ra các phân tích và dự báo hợp lý.

```r
 # Đọc dữ liệu
  df <- read_csv("ASEAN_2020_2023_Final_Cleaned.csv")

  # Chuyển đổi dữ liệu số (loại bỏ cột Quốc_gia nếu có)
  df_numeric <- df %>% select(-Quốc_gia)

  # Tính toán ma trận tương quan
  cor_matrix <- cor(df_numeric, use = "complete.obs")

  # Vẽ heatmap với corrplot
  corrplot(cor_matrix, method = "color", type = "upper",
           col = colorRampPalette(c("blue", "white", "red"))(200),
           tl.col = "black", tl.srt = 45, addCoef.col = "black")

  # Chuyển đổi ma trận tương quan sang dạng dữ liệu dài
  cor_melted <- melt(cor_matrix)

  # Vẽ heatmap với ggplot2
  ggplot(cor_melted, aes(x = Var1, y = Var2, fill = value)) +
    geom_tile(color = "white") +
    scale_fill_gradient2(low = "blue", mid = "white", high = "red", midpoint = 0) +
    labs(title = "Heatmap: Mối Tương Quan Giữa Các Biến", x = "", y = "") +
    theme_minimal() +
    theme(axis.text.x = element_text(angle = 45, hjust = 1))
```

*   Các hệ số tương quan âm (khoảng -0.3) giữa dân số và GDP có thể chỉ ra rằng khi dân số tăng, GDP không nhất thiết tăng theo cùng tỷ lệ hoặc có thể bị ảnh hưởng tiêu cực trong một số trường hợp.

##Thực hiện phân cụm K-Means
- Để thực hiện phân cụm K-Means dựa trên các yếu tố kinh tế – xã hội, nhóm tiến hành tổng hợp dữ liệu từ ba nguồn: GDP, dân số và tỷ lệ thất nghiệp của các quốc gia ASEAN trong giai đoạn 2020–2023.
- Sau khi gộp dữ liệu, nhóm xử lý các giá trị thiếu, chuẩn hóa dữ liệu, và áp dụng thuật toán K-Means để phân nhóm các quốc gia thành 3 cụm chính dựa trên mức độ tương đồng về đặc điểm.
- Kết quả phân cụm sẽ được sử dụng cho việc trực quan hóa và phân tích xu hướng phát triển của từng cụm quốc gia

```r
# Cài đặt gói nếu cần
install.packages("tidyverse")
install.packages("cluster")
install.packages("factoextra")

# Gọi thư viện
library(tidyverse)
library(cluster)
library(factoextra)

# Đọc dữ liệu từ các file CSV (sau khi upload lên Google Colab)
ty_le_that_nghiep <- read_csv("ty_le_that_nghiep_cleaned.csv")
dan_so_2020_2023 <- read_csv("dan_so_2020_2023_fixed.csv")
GDP_ASEAN_2020_2023 <- read_csv("GDP_ASEAN_2020_2023_Cleaned.csv")

# Đổi tên cột "Country" thành "Quốc gia" để thống nhất
ty_le_that_nghiep <- ty_le_that_nghiep %>%
  rename("Quốc gia" = 1)

dan_so_2020_2023 <- dan_so_2020_2023 %>%
  rename("Quốc gia" = 1)

GDP_ASEAN_2020_2023 <- GDP_ASEAN_2020_2023 %>%
  rename("Quốc gia" = 1)

# Kết hợp 3 bảng dữ liệu lại dựa trên "Quốc gia"
data_combined <- ty_le_that_nghiep %>%
  left_join(dan_so_2020_2023, by = "Quốc gia") %>%
  left_join(GDP_ASEAN_2020_2023, by = "Quốc gia")

# Thay thế các giá trị NA bằng trung bình theo cột
data_combined[, 2:ncol(data_combined)] <- data_combined[, 2:ncol(data_combined)] %>%
  mutate(across(everything(), ~ ifelse(is.na(.), mean(., na.rm = TRUE), .)))

# Loại bỏ cột "Quốc gia" để chuẩn bị dữ liệu phân cụm
data_for_clustering <- data_combined %>%
  select(-`Quốc gia`)

# Chuẩn hóa dữ liệu
data_for_clustering_scaled <- scale(data_for_clustering)

# Kiểm tra dữ liệu chuẩn hóa
print(head(data_for_clustering_scaled))

# Phân cụm K-Means (dùng khoảng cách Euclidean mặc định)
set.seed(123)
kmeans_result <- kmeans(data_for_clustering_scaled, centers = 3, nstart = 25)

# Thêm kết quả phân cụm vào bảng gốc
data_combined$cluster <- kmeans_result$cluster

# Xem kết quả phân cụm
head(data_combined)


```

##Thực hiện phân cụm Hierarchical Clustering và vẽ Dendrogram

- Trước khi thực hiện phân tích và phân cụm, nhóm tiến hành tổng hợp dữ liệu từ ba nguồn: tỷ lệ thất nghiệp, dân số và GDP của các quốc gia Đông Nam Á trong giai đoạn 2020–2023.

- Dữ liệu được gộp lại theo cột "Quốc gia" để tạo thành một bảng duy nhất (data_combined). Sau đó, nhóm tiến hành làm sạch dữ liệu bằng cách thay thế các giá trị bị thiếu (NA) bằng giá trị trung bình của từng cột – giúp đảm bảo dữ liệu đầy đủ và sẵn sàng cho phân tích tiếp theo

```r
#Cài đặt gói nếu cần
install.packages("tidyverse")
install.packages("cluster")
install.packages("factoextra")

# Gọi thư viện
library(tidyverse)
library(cluster)
library(factoextra)

# Đọc dữ liệu từ các file CSV (sau khi upload lên Google Colab)
ty_le_that_nghiep <- read_csv("ty_le_that_nghiep_cleaned.csv")
dan_so_2020_2023 <- read_csv("dan_so_2020_2023_fixed.csv")
GDP_ASEAN_2020_2023 <- read_csv("GDP_ASEAN_2020_2023_Cleaned.csv")

# Đổi tên cột "Country" thành "Quốc gia" để thống nhất
ty_le_that_nghiep <- ty_le_that_nghiep %>%
  rename("Quốc gia" = 1)

dan_so_2020_2023 <- dan_so_2020_2023 %>%
  rename("Quốc gia" = 1)

GDP_ASEAN_2020_2023 <- GDP_ASEAN_2020_2023 %>%
  rename("Quốc gia" = 1)

# Kết hợp 3 bảng dữ liệu lại dựa trên "Quốc gia"
data_combined <- ty_le_that_nghiep %>%
  left_join(dan_so_2020_2023, by = "Quốc gia") %>%
  left_join(GDP_ASEAN_2020_2023, by = "Quốc gia")

# Thay thế các giá trị NA bằng trung bình theo cột
data_combined[, 2:ncol(data_combined)] <- data_combined[, 2:ncol(data_combined)] %>%
  mutate(across(everything(), ~ ifelse(is.na(.), mean(., na.rm = TRUE), .)))
```

- Để hiểu rõ hơn mối quan hệ giữa các quốc gia trong khu vực Đông Nam Á, nhóm thực hiện phân cụm thứ bậc (Hierarchical Clustering).
- Phương pháp này giúp xây dựng cây phân cấp (dendrogram), qua đó có thể quan sát trực quan sự tương đồng giữa các quốc gia dựa trên các chỉ số kinh tế – xã hội như GDP, dân số và tỷ lệ thất nghiệp

```r
# Chuẩn hóa dữ liệu
data_for_clustering <- data_combined %>% select(-`Quốc gia`)
data_scaled <- scale(data_for_clustering)

# Tính khoảng cách Euclidean
distance_matrix <- dist(data_scaled, method = "euclidean")

# Áp dụng thuật toán phân cụm theo Ward.D2
hc_result <- hclust(distance_matrix, method = "ward.D2")

# Vẽ biểu đồ Dendrogram
plot(hc_result, labels = data_combined$`Quốc gia`, main = "Dendrogram PHAN CUM THU BAC", xlab = "", sub = "")
rect.hclust(hc_result, k = 3, border = "red")  # Cắt cây thành 3 cụm
```

```r
# BƯỚC 1: Chuẩn bị dữ liệu
prepare_data <- function(df, year) {
  cat("\n🔹 BƯỚC 1: Chuẩn bị dữ liệu cho năm", year, "\n")

  # Xác định cột cơ bản
  population_col <- paste0("Dan_so_", year, ".x")
  gdp_col <- paste0("GDP_", year, ".x")
  unemployment_col <- paste0("X", year, ".x")

  # Với năm từ 2021 trở đi thì mới có cột tăng trưởng dân số
  if (year > 2020) {
    growth_col <- paste0("Tang_truong_Dan_so_", year)
    required_cols <- c("Quốc_gia", population_col, growth_col, gdp_col, unemployment_col)
  } else {
    required_cols <- c("Quốc_gia", population_col, gdp_col, unemployment_col)
  }

  # Kiểm tra các cột có tồn tại
  missing_cols <- setdiff(required_cols, colnames(df))
  if (length(missing_cols) > 0) {
    stop(paste("❌ Thiếu cột:", paste(missing_cols, collapse = ", ")))
  }

  # Chọn các cột cần thiết cho năm
  data_year <- df[, required_cols]
  cat("✅ Dữ liệu cho năm", year, "đã được chuẩn bị thành công.\n")

  # In kết quả dữ liệu cho năm, đảm bảo in đầy đủ 11 quốc gia
  print(data_year)  # In toàn bộ dữ liệu của năm

  return(data_year)
}

# Gọi hàm để chuẩn bị dữ liệu cho tất cả các năm từ 2020 đến 2023
data_2020 <- prepare_data(df, 2020)
data_2021 <- prepare_data(df, 2021)
data_2022 <- prepare_data(df, 2022)
data_2023 <- prepare_data(df, 2023)
```

###Tính khoảng cách Euclidean

Bước 2: Tính khoảng cách Euclidean giữa các quốc gia dựa trên các chỉ số đã được chuẩn bị ở bước 1.

*   Chuẩn hóa dữ liệu: Dữ liệu được chuẩn hóa (scaled) để các giá trị của các chỉ số như dân số, GDP, tỷ lệ thất nghiệp có thể so sánh được với nhau. Hàm scale() sẽ chuẩn hóa tất cả các cột ngoại trừ cột Quốc_gia.
*   Tính khoảng cách Euclidean: Khoảng cách Euclidean giữa các quốc gia được tính bằng hàm dist(). Phương thức "euclidean" sẽ tính toán khoảng cách giữa các điểm trong không gian đa chiều.

*   In kết quả: Ma trận khoảng cách Euclidean sẽ được in ra nếu dữ liệu không quá lớn. Ma trận này cho thấy khoảng cách giữa các quốc gia, giúp đánh giá độ tương đồng giữa các quốc gia dựa trên các chỉ số.





```r
# BƯỚC 2: Tính khoảng cách Euclidean
calculate_distance <- function(data_year) {
  cat("\n🔹 BƯỚC 2: Tính khoảng cách Euclidean\n")

  data_scaled <- scale(data_year[, -1])  # Loại bỏ cột 'Quốc_gia'
  rownames(data_scaled) <- data_year$`Quốc_gia`
  dist_matrix <- dist(data_scaled, method = "euclidean")

  cat("✅ Khoảng cách Euclidean đã được tính toán thành công.\n")

  # In ra khoảng cách (chỉ in ma trận nếu dữ liệu không quá lớn)
  print(as.matrix(dist_matrix))  # In ma trận khoảng cách Euclidean

  return(dist_matrix)
}

```

Khi bạn gọi hàm calculate_distance(data_2020), nó sẽ thực hiện các bước sau:

*   Chuẩn hóa dữ liệu: Dữ liệu từ năm 2020 sẽ được chuẩn hóa (loại bỏ cột "Quốc_gia" và chuẩn hóa các giá trị còn lại)
*   Tính khoảng cách Euclidean: Hàm sẽ tính toán khoảng cách Euclidean giữa các quốc gia dựa trên các chỉ số đã chuẩn hóa.

*   In ra ma trận khoảng cách: Ma trận khoảng cách Euclidean giữa các quốc gia sẽ được in ra trong console, giúp hiểu rõ độ tương đồng giữa các quốc gia theo các chỉ số đã chọn.

Kết quả:
Sau khi gọi hàm dist_matrix_2020 <- calculate_distance(data_2020), bạn sẽ nhận được một ma trận khoảng cách Euclidean giữa các quốc gia trong dữ liệu năm 2020.

###Phân cụm và vẽ Dendrogram

Hàm plot_dendrogram() yêu cầu thực hiện phân cụm và vẽ dendrogram cho dữ liệu của từng năm. Dưới đây là giải thích về nội dung và kết quả của hàm:

1.   Nội dung:

*   Phân cụm Hierarchical (hclust): Sử dụng phương pháp ward.D2 để thực hiện phân cụm dữ liệu theo khoảng cách Euclidean. Phương pháp Ward là một trong các phương pháp phân cụm phổ biến, tối ưu hóa sự đồng nhất trong các nhóm.
*   Vẽ Dendrogram: Sau khi phân cụm, một dendrogram (biểu đồ phân cụm cây) được vẽ để thể hiện mối quan hệ giữa các quốc gia, dựa trên các đặc trưng như dân số, GDP, tăng trưởng dân số, và tỷ lệ thất nghiệp trong năm cụ thể.

*   Tên và nhãn cho Dendrogram: Mỗi dendrogram có tiêu đề là "Hierarchical Clustering - Năm [Năm]", và các quốc gia được hiển thị trên trục hoành.





```r
# BƯỚC 3: Phân cụm và vẽ Dendrogram
plot_dendrogram <- function(dist_matrix, year) {
  cat("\n🔹 BƯỚC 3: Phân cụm và vẽ Dendrogram cho năm", year, "\n")

  hc <- hclust(dist_matrix, method = "ward.D2")

  # Vẽ dendrogram cho từng năm riêng biệt
  plot(hc,
       main = paste("Hierarchical Clustering - Năm", year),
       xlab = "", sub = "", cex = 0.8)

  cat("✅ Dendrogram cho năm", year, "đã được vẽ thành công.\n")
}

```

Biểu đồ dendrogram sẽ hiển thị sự phân nhóm các quốc gia ASEAN dựa trên các chỉ số liên quan đến dân số, GDP, tỷ lệ thất nghiệp trong năm 2020. Biểu đồ này sẽ giúp hình dung các quốc gia tương tự nhau về mặt các chỉ số, với các nhánh của cây biểu diễn các nhóm quốc gia có sự tương đồng cao.

Hàm R (perform_clustering) thực hiện phân cụm thứ bậc theo từng năm (2020–2023).

*   Chọn biến phân cụm: Dân số, GDP, tỷ lệ thất nghiệp, và nếu > 2020 thì thêm tốc độ tăng trưởng dân số.
*   Chuẩn hóa dữ liệu: Đảm bảo các biến cùng thang đo.

*   Tính khoảng cách: Dùng khoảng cách Euclidean.
*   Phân cụm: Dùng phương pháp Ward, cắt thành 3 cụm.

*   Xuất kết quả: Gán nhãn cụm vào dataframe và in ra Quốc gia + Cụm tương ứng.








```r
# BƯỚC 4: Phân cụm theo các năm (2020-2023)
perform_clustering <- function(df, year) {
  cat("\n🔹 BƯỚC 4: Phân cụm cho năm", year, "\n")

  # Chọn các cột cần thiết cho phân cụm (dân số, GDP và tỷ lệ thất nghiệp)
  if (year > 2020) {
    growth_col <- paste0("Tang_truong_Dan_so_", year)
    clustering_data <- df[, c("Quốc_gia", paste0("Dan_so_", year, ".x"), paste0("GDP_", year, ".x"), paste0("X", year, ".x"), growth_col)]
  } else {
    clustering_data <- df[, c("Quốc_gia", paste0("Dan_so_", year, ".x"), paste0("GDP_", year, ".x"), paste0("X", year, ".x"))]
  }

  # Chuẩn hóa dữ liệu
  data_scaled <- scale(clustering_data[, -1])  # Loại bỏ cột "Quốc_gia"
  rownames(data_scaled) <- clustering_data$`Quốc_gia`

  # Tính toán ma trận khoảng cách Euclidean
  dist_matrix <- dist(data_scaled, method = "euclidean")

  # Phân cụm với phương pháp Ward
  hc <- hclust(dist_matrix, method = "ward.D2")

  # Cắt cây phân cụm thành 3 nhóm
  groups <- cutree(hc, k = 3)

  # Thêm nhóm phân cụm vào dataframe
  df$Cluster <- groups

  # In kết quả phân cụm (Quốc gia và nhóm phân cụm)
  cat("✅ Phân cụm cho năm", year, "hoàn thành.\n")
  print(df[, c("Quốc_gia", "Cluster")])  # In quốc gia và nhóm phân cụm

  return(df)
}


```

# Xác định cụm -	Xác định số cụm tối ưu
- Phương pháp Elbow Method(Khuỷu tay)
  + Với mỗi giá trị 𝑘, thuật toán K-Means tính WSS (Within-cluster Sum of Squares) – tổng khoảng cách bình phương từ mỗi điểm đến tâm cụm của nó.
  + Khi số cụm tăng, WSS giảm, nhưng đến một điểm "gãy khuỷu tay", việc tăng cụm không còn giúp cải thiện WSS nhiều nữa.
  + Điểm này chính là số cụm tối
- Silhouette Score
  + Đo độ “tách biệt” và “gắn kết” của cụm:
    + Gần 1 → phân cụm rõ, chất lượng tốt.
    + Gần 0 → các cụm chồng lấn, phân cụm không rõ ràng
- Chọn k phù hợp rất quan trọng.
  + Dùng Elbow Method: xem điểm "gãy" trong biểu đồ SSE (tổng khoảng cách bình phương).
  + Dùng Silhouette Score: đánh giá độ tách biệt giữa các cụm.
-> Nhận xét: Việc sử dụng kết hợp hai phương pháp xác định số cụm giúp tăng độ tin cậy. Đây là một bước cực kỳ quan trọng để tránh phân cụm quá ít (chung chung) hoặc quá nhiều (nhiễu loạn).
